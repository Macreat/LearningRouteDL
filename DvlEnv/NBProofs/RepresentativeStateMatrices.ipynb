# generate 2 transition matrices sizing N x N and plot in one raw thier heatmaps

# compute the vector holding the sum for each colum, plot below the heat map of each vector, and compute the joint matrix between the two transition matrices

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_transition_matrix(N):
    """
    Generate a random N x N transition matrix.
    Each row will sum to 1 to represent probabilities.
    """
    matrix = np.random.rand(N, N)  # Generate a random NxN matrix with values [0, 1)
    matrix /= matrix.sum(axis=1, keepdims=True)  # Normalize rows to sum to 1
    return matrix

# Define the size of the matrices
N = 16  # Replace with desired size

# Generate two transition matrices
transition_matrix_1 = generate_transition_matrix(N)
transition_matrix_2 = generate_transition_matrix(N)

# Compute column sums for each transition matrix
column_sums_1 = transition_matrix_1.sum(axis=0)  # Column-wise sum for matrix 1
column_sums_2 = transition_matrix_2.sum(axis=0)  # Column-wise sum for matrix 2

# Plot heatmaps and column sum vectors
fig, axes = plt.subplots(3, 2, figsize=(12, 15))  # 3 rows and 2 columns

# Heatmap for Transition Matrix 1
im1 = axes[0, 0].imshow(transition_matrix_1, cmap="viridis", aspect="auto")
axes[0, 0].set_title("Transition Matrix 1 Heatmap")
fig.colorbar(im1, ax=axes[0, 0])
axes[0, 0].set_xlabel("To State")
axes[0, 0].set_ylabel("From State")

# Bar plot for Column Sums of Transition Matrix 1
axes[1, 0].bar(range(N), column_sums_1/column_sums_1.sum(axis=0), color="blue", alpha=0.7)
axes[1, 0].set_title("Column Sums of Transition Matrix 1")
axes[1, 0].set_xlabel("Column Index")
axes[1, 0].set_ylabel("Sum")

# Heatmap for Transition Matrix 2
im2 = axes[0, 1].imshow(transition_matrix_2, cmap="viridis", aspect="auto")
axes[0, 1].set_title("Transition Matrix 2 Heatmap")
fig.colorbar(im2, ax=axes[0, 1])
axes[0, 1].set_xlabel("To State")
axes[0, 1].set_ylabel("From State")

# Bar plot for Column Sums of Transition Matrix 2
axes[1, 1].bar(range(N), column_sums_2/column_sums_2.sum(axis=0), color="green", alpha=0.7)
axes[1, 1].set_title("Column Sums of Transition Matrix 2")
axes[1, 1].set_xlabel("Column Index")
axes[1, 1].set_ylabel("Sum")

# Compute eigenvalues for each transition matrix
eigenvalues_1 = np.linalg.eigvals(transition_matrix_1)
axes[2, 0].bar(range(len(eigenvalues_1)), eigenvalues_1.real, color="blue", alpha=0.7)
axes[2, 0].set_title("Eigenvalues of Transition Matrix 1")
axes[2, 0].set_xlabel("Eigenvalue Index")
axes[2, 0].set_ylabel("Real Part")

eigenvalues_2 = np.linalg.eigvals(transition_matrix_2)
axes[2, 1].bar(range(len(eigenvalues_2)), eigenvalues_2.real, color="green", alpha=0.7)
axes[2, 1].set_title("Eigenvalues of Transition Matrix 2")
axes[2, 1].set_xlabel("Eigenvalue Index")
axes[2, 1].set_ylabel("Real Part")

plt.tight_layout()
plt.show()

# Compute joint transition matrix
joint_matrix = np.dot(transition_matrix_1, transition_matrix_2)  # Matrix multiplication

# Compute column sums for the joint matrix
column_sums_joint = joint_matrix.sum(axis=0)

# Compute eigenvalues for the joint matrix
eigenvalues_joint = np.linalg.eigvals(joint_matrix)

# Plot heatmap, column sums, and eigenvalues for the joint matrix
fig, axes = plt.subplots(3, 1, figsize=(10, 15))  # 3 rows, 1 column

# Heatmap for Joint Transition Matrix
im = axes[0].imshow(joint_matrix, cmap="viridis", aspect="auto")
axes[0].set_title("Joint Transition Matrix Heatmap")
fig.colorbar(im, ax=axes[0])
axes[0].set_xlabel("To State")
axes[0].set_ylabel("From State")

# Bar plot for Column Sums of Joint Transition Matrix
axes[1].bar(range(N), column_sums_joint/column_sums_joint.sum(axis=0), color="blue", alpha=0.7)
axes[1].set_title("Column Sums of Joint Transition Matrix")
axes[1].set_xlabel("Column Index")
axes[1].set_ylabel("Sum")

# Bar plot for Eigenvalues of Joint Transition Matrix
axes[2].bar(range(len(eigenvalues_joint)), eigenvalues_joint.real, color="green", alpha=0.7)
axes[2].set_title("Eigenvalues of Joint Transition Matrix")
axes[2].set_xlabel("Eigenvalue Index")
axes[2].set_ylabel("Real Part")

plt.tight_layout()
plt.show()

################################################################################
#### compute the conditional transition matrix between bot transtition matrices
################################################################################

# Generate two addionally transition matrices
transition_matrix_3 = generate_transition_matrix(N)
transition_matrix_4 = generate_transition_matrix(N)

# Compute the conditional transition matrix
conditional_matrix = transition_matrix_3 * transition_matrix_4  # Element-wise product
conditional_matrix /= conditional_matrix.sum(axis=1, keepdims=True)  # Normalize rows to sum to 1

# Compute column sums for the conditional matrix
column_sums_conditional = conditional_matrix.sum(axis=0)

# Compute eigenvalues for the conditional matrix
eigenvalues_conditional = np.linalg.eigvals(conditional_matrix)

# Plot heatmap, column sums, and eigenvalues for the conditional matrix
fig, axes = plt.subplots(3, 1, figsize=(12, 15))  # 3 rows, 1 column

# Heatmap for Conditional Transition Matrix
im = axes[0].imshow(conditional_matrix, cmap="viridis", aspect="auto")
axes[0].set_title("Conditional Transition Matrix Heatmap")
fig.colorbar(im, ax=axes[0])
axes[0].set_xlabel("To State")
axes[0].set_ylabel("From State")

# Bar plot for Column Sums of Conditional Transition Matrix
axes[1].bar(range(N), column_sums_conditional/column_sums_joint.sum(axis=0), color="blue", alpha=0.7)
axes[1].set_title("Column Sums of Conditional Transition Matrix")
axes[1].set_xlabel("Column Index")
axes[1].set_ylabel("Sum")

# Bar plot for Eigenvalues of Conditional Transition Matrix
axes[2].bar(range(len(eigenvalues_conditional)), eigenvalues_conditional.real, color="green", alpha=0.7)
axes[2].set_title("Eigenvalues of Conditional Transition Matrix")
axes[2].set_xlabel("Eigenvalue Index")
axes[2].set_ylabel("Real Part")

plt.tight_layout()
plt.show()
```

# complete conversion of NumPy-based transition matrix analysis to PyTorch:

```python
import torch
import matplotlib.pyplot as plt

def generate_transition_matrix(N, device='cpu'):
    """
    Generate a random N x N transition matrix in PyTorch.
    Each row will sum to 1 to represent probabilities.
    """
    matrix = torch.rand(N, N, device=device)  # Generate random matrix on specified device
    matrix /= matrix.sum(dim=1, keepdim=True)  # Normalize rows to sum to 1
    return matrix

# Set device and matrix size
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
N = 16  # Number of states

# Generate two transition matrices
transition_matrix_1 = generate_transition_matrix(N, device)
transition_matrix_2 = generate_transition_matrix(N, device)

# Compute column sums for each transition matrix
column_sums_1 = transition_matrix_1.sum(dim=0)  # Column-wise sum for matrix 1
column_sums_2 = transition_matrix_2.sum(dim=0)  # Column-wise sum for matrix 2

# Plot heatmaps and column sum vectors
fig, axes = plt.subplots(3, 2, figsize=(12, 15))

# Heatmap for Transition Matrix 1
im1 = axes[0, 0].imshow(transition_matrix_1.cpu().numpy(), cmap="viridis", aspect="auto")
axes[0, 0].set_title("Transition Matrix 1 Heatmap")
fig.colorbar(im1, ax=axes[0, 0])
axes[0, 0].set_xlabel("To State")
axes[0, 0].set_ylabel("From State")

# Bar plot for Column Sums of Transition Matrix 1
axes[1, 0].bar(range(N), (column_sums_1/column_sums_1.sum()).cpu().numpy(), color="blue", alpha=0.7)
axes[1, 0].set_title("Column Sums of Transition Matrix 1")
axes[1, 0].set_xlabel("Column Index")
axes[1, 0].set_ylabel("Sum")

# Heatmap for Transition Matrix 2
im2 = axes[0, 1].imshow(transition_matrix_2.cpu().numpy(), cmap="viridis", aspect="auto")
axes[0, 1].set_title("Transition Matrix 2 Heatmap")
fig.colorbar(im2, ax=axes[0, 1])
axes[0, 1].set_xlabel("To State")
axes[0, 1].set_ylabel("From State")

# Bar plot for Column Sums of Transition Matrix 2
axes[1, 1].bar(range(N), (column_sums_2/column_sums_2.sum()).cpu().numpy(), color="green", alpha=0.7)
axes[1, 1].set_title("Column Sums of Transition Matrix 2")
axes[1, 1].set_xlabel("Column Index")
axes[1, 1].set_ylabel("Sum")

# Compute eigenvalues for each transition matrix
eigenvalues_1 = torch.linalg.eigvals(transition_matrix_1)
axes[2, 0].bar(range(len(eigenvalues_1)), eigenvalues_1.real.cpu().numpy(), color="blue", alpha=0.7)
axes[2, 0].set_title("Eigenvalues of Transition Matrix 1")
axes[2, 0].set_xlabel("Eigenvalue Index")
axes[2, 0].set_ylabel("Real Part")

eigenvalues_2 = torch.linalg.eigvals(transition_matrix_2)
axes[2, 1].bar(range(len(eigenvalues_2)), eigenvalues_2.real.cpu().numpy(), color="green", alpha=0.7)
axes[2, 1].set_title("Eigenvalues of Transition Matrix 2")
axes[2, 1].set_xlabel("Eigenvalue Index")
axes[2, 1].set_ylabel("Real Part")

plt.tight_layout()
plt.show()

# Compute joint transition matrix (matrix multiplication)
joint_matrix = torch.mm(transition_matrix_1, transition_matrix_2)

# Compute column sums for the joint matrix
column_sums_joint = joint_matrix.sum(dim=0)

# Compute eigenvalues for the joint matrix
eigenvalues_joint = torch.linalg.eigvals(joint_matrix)

# Plot heatmap, column sums, and eigenvalues for the joint matrix
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Heatmap for Joint Transition Matrix
im = axes[0].imshow(joint_matrix.cpu().numpy(), cmap="viridis", aspect="auto")
axes[0].set_title("Joint Transition Matrix Heatmap")
fig.colorbar(im, ax=axes[0])
axes[0].set_xlabel("To State")
axes[0].set_ylabel("From State")

# Bar plot for Column Sums of Joint Transition Matrix
axes[1].bar(range(N), (column_sums_joint/column_sums_joint.sum()).cpu().numpy(), color="blue", alpha=0.7)
axes[1].set_title("Column Sums of Joint Transition Matrix")
axes[1].set_xlabel("Column Index")
axes[1].set_ylabel("Sum")

# Bar plot for Eigenvalues of Joint Transition Matrix
axes[2].bar(range(len(eigenvalues_joint)), eigenvalues_joint.real.cpu().numpy(), color="green", alpha=0.7)
axes[2].set_title("Eigenvalues of Joint Transition Matrix")
axes[2].set_xlabel("Eigenvalue Index")
axes[2].set_ylabel("Real Part")

plt.tight_layout()
plt.show()

################################################################################
#### Compute the conditional transition matrix between both transition matrices
################################################################################

# Generate two additional transition matrices
transition_matrix_3 = generate_transition_matrix(N, device)
transition_matrix_4 = generate_transition_matrix(N, device)

# Compute the conditional transition matrix (element-wise product)
conditional_matrix = transition_matrix_3 * transition_matrix_4  # Hadamard product
conditional_matrix /= conditional_matrix.sum(dim=1, keepdim=True)  # Normalize rows

# Compute column sums for the conditional matrix
column_sums_conditional = conditional_matrix.sum(dim=0)

# Compute eigenvalues for the conditional matrix
eigenvalues_conditional = torch.linalg.eigvals(conditional_matrix)

# Plot heatmap, column sums, and eigenvalues for the conditional matrix
fig, axes = plt.subplots(3, 1, figsize=(12, 15))

# Heatmap for Conditional Transition Matrix
im = axes[0].imshow(conditional_matrix.cpu().numpy(), cmap="viridis", aspect="auto")
axes[0].set_title("Conditional Transition Matrix Heatmap")
fig.colorbar(im, ax=axes[0])
axes[0].set_xlabel("To State")
axes[0].set_ylabel("From State")

# Bar plot for Column Sums of Conditional Transition Matrix
axes[1].bar(range(N), (column_sums_conditional/column_sums_conditional.sum()).cpu().numpy(), color="blue", alpha=0.7)
axes[1].set_title("Column Sums of Conditional Transition Matrix")
axes[1].set_xlabel("Column Index")
axes[1].set_ylabel("Sum")

# Bar plot for Eigenvalues of Conditional Transition Matrix
axes[2].bar(range(len(eigenvalues_conditional)), eigenvalues_conditional.real.cpu().numpy(), color="green", alpha=0.7)
axes[2].set_title("Eigenvalues of Conditional Transition Matrix")
axes[2].set_xlabel("Eigenvalue Index")
axes[2].set_ylabel("Real Part")

plt.tight_layout()
plt.show()
```
